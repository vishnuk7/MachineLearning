### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for ploting in juptyer
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x=[1,2,3,4,5]
y=[11,22,33,44,55]

# 2. setup plot
fig,ax = plt.subplots()

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(
    title='Simple Plot',
    xlabel='x-axis',
    ylabel='y-axis'
)

# 5.  save and show (you save the whole figure)
fig.savefig('img/sample.png')
```
