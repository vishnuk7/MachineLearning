### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for ploting in juptyer
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x=[1,2,3,4,5]
y=[11,22,33,44,55]

# 2. setup plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(
    title='Simple Plot',
    xlabel='x-axis',
    ylabel='y-axis'
)

# 5.  save and show (you save the whole figure)
fig.savefig('img/sample.png')
```

#### Two options for subplots 

```python
#subplots options 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, 
                                           ncols=2,
                                           figsize=(10,5)
                                          )
#plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

why we are using `3rd` method over methods?

The main reason using above command is that we can use one command to lots of  figure 

```python
#subplot option 2

fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5)
                     )

#plot to each different axis
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

#### which one should you use? (pyplot vs matplotlib OO method)

- when ploting something quickly, okay to use the pyplot method  
- whne ploting something advance then use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

### pyplot method

```python
over_50.plot(kind="scatter",x='age',y="chol",c='target');
```

### OO method  mixed with pyplot

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
# ax.set_xlim([45,100]);
```

### OO method from scratch

```python
fig,ax = plt.subplots(figsize = (10,6))

#plot the data 
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]
                    );

#customize the plot
ax.set(title='Heart Disease and Cholesterol',
       xlabel='Age',
       ylabel='Cholesterol'
      );

#add legend
ax.legend(*scatter.legend_elements(), title='target');

#add horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
over_50.head()
```

```python
#subplots of chol, thalach, age
fig,(ax0,ax1) = plt.subplots(figsize = (10,10),
                            nrows=2,
                            ncols=1,
                             sharex=True);

#plot the data
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"]);

#customize the plot
ax0.set(title='Heart Disease and Cholesterol Levels',
       ylabel='Cholesterol'
      );

#add legend
ax0.legend(*scatter.legend_elements(), title='target');

#add horizontal line
ax0.axhline(over_50["chol"].mean(),linestyle="--");


#plot the data
scatter=ax1.scatter(x=over_50["age"], 
                    y=over_50["thalach"],
                    c=over_50["target"]);

#customize the plot
ax1.set(title='Heart Disease and Max heart rate',
       xlabel='Age',
       ylabel='Max heart rate'
      );

#add legend
ax1.legend(*scatter.legend_elements(), title='target');

#add horizontal line
ax1.axhline(over_50["thalach"].mean(),linestyle="--");

#add title on figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```

```python
#set the style
plt.style.use('seaborn-whitegrid')

fig,ax = plt.subplots(figsize = (10,6))

#plot the data 
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter" #change the color scheme
                    );

#customize the plot
ax.set(title='Heart Disease and Cholesterol',
       xlabel='Age',
       ylabel='Cholesterol'
      );

#add legend
ax.legend(*scatter.legend_elements(), title='target');

#add horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
#customizing the x-axis and y-axis limitations
#subplots of chol, thalach, age
fig,(ax0,ax1) = plt.subplots(figsize = (10,10),
                            nrows=2,
                            ncols=1,
                             sharex=True);

#plot the data
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter");

#customize the plot
ax0.set(title='Heart Disease and Cholesterol Levels',
       ylabel='Cholesterol'
      );

#add legend
ax0.legend(*scatter.legend_elements(), title='target');

#add horizontal line
ax0.axhline(over_50["chol"].mean(),linestyle="--");


#change the x axis limits
ax0.set_xlim([50,80])

#plot the data
scatter=ax1.scatter(x=over_50["age"], 
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter");

#customize the plot
ax1.set(title='Heart Disease and Max heart rate',
       xlabel='Age',
       ylabel='Max heart rate'
      );

#add legend
ax1.legend(*scatter.legend_elements(), title='target');

#add y axis limit
ax1.set_ylim([60,200])

#add horizontal line
ax1.axhline(over_50["thalach"].mean(),linestyle="--");



#add title on figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
