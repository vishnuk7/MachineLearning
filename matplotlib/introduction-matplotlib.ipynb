### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for ploting in juptyer
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x=[1,2,3,4,5]
y=[11,22,33,44,55]

# 2. setup plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(
    title='Simple Plot',
    xlabel='x-axis',
    ylabel='y-axis'
)

# 5.  save and show (you save the whole figure)
fig.savefig('img/sample.png')
```

### Making figures using numpy arrays

#### Line plot

```python
import numpy as np
```

```python
#create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
#plot the data
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

#### Scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
#another scatter plot 
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
#Make a plot from dictionary
nut_butter_prices = {
    "Almond butter": 10,
    "Peanut butter": 8,
    "Cashew butter":12
}
```

```python
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="NUT BUTTER", ylabel="Pri")
```
