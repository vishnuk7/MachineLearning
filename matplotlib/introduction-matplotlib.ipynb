### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for ploting in juptyer
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x=[1,2,3,4,5]
y=[11,22,33,44,55]

# 2. setup plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(
    title='Simple Plot',
    xlabel='x-axis',
    ylabel='y-axis'
)

# 5.  save and show (you save the whole figure)
fig.savefig('img/sample.png')
```

```python
#Make a plot from dictionary
nut_butter_prices = {
    "Almond butter": 10,
    "Peanut butter": 8,
    "Cashew butter":12
}
```

```python
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="NUT BUTTER", ylabel="Prices ($)");
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

#### Histograms 

```python
#make some data 

x=np.random.randn(100)

#plot histogram
fig, ax = plt.subplots()
ax.hist(x);
```

#### Two options for subplots 

```python
#subplots options 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, 
                                           ncols=2,
                                           figsize=(10,5)
                                          )
#plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

why we are using `3rd` method over methods?

The main reason using above command is that we can use one command to lots of  figure 

```python
#subplot option 2

fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5)
                     )

#plot to each different axis
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```
